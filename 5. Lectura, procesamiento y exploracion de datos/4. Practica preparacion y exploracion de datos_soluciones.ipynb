## 1.-

```python
import pandas as pd
import numpy as np

data=pd.read_excel('datos/all-euro-data-2020-2021.xlsx',sheet_name='SP1')
```

## 2.-

```python
print(data.head(100))
data=data.fillna(data.median())
data.head(100)
```

En este caso los **datos venian bastante limpios** y no hay valores nulos practicamente.

## 3.-

```python
feats=['HTHG','HTAG','FTHG','FTAG']
print('Valores medios:\n',data[feats].mean())
print()
print('Valores medios:\n',data[feats].std())
```

## 4.-

```python
data[feats].corr()
```

Las **parejas más correlacionadas** (linealmente y positivamente) son: **(FTAG,HTAG) y (FTHG,HTHG)**:

 - El número de **goles** (tanto para el equipo de casa como para el visitante) marcados **en el descanso** están fuertemente (y positivamente) **correlacionados con el total de goles marcados al final** del partido (es **sentido común confirmado cuantitativamente**).
 
 - Los **goles marcados por los dos equipos (home/away)** están **correlados negativamente** (cuando una cantidad crece la otra decrece).

## 5.-

```python
covarianza=data[feats].cov()
print(covarianza)
print()
varianza=[]
for feat in feats:
    for indice in covarianza.index.values:
        if feat==indice:
            varianza.append(covarianza[feat].loc[indice])
    
varianza
```

```python
varianza=np.array(varianza)
print(varianza)
varianza=np.sort(varianza)
varianza
```

**'FTHG' y 'FTAG' son las variables que APARENTEMENTE tienen mayor variabilidad (en ese orden)**. **Más tarde se verá lo que ocurre con estas magnitudes cuando las variables estén normalizadas.**

   - Los **valores fuera de la diagonal** indican **como varía una variable cuando crece su pareja** en la matriz.

   - **Valores positivos** indican que **ambas variables crecen o decrecen a la vez**.
    
   - **Valores negativos** indican que ambas variables **se mueven en direcciones contrarias** (cuando una crece la otra decrece)
    
   - Es importante tener en cuenta que las **variables no están normalizadas** por lo que en general los **valores absolutos** para distintas parejas **no son en general comparables** con otras parejas.
   
   - En este caso concreto dado que **todas las variables estudiadas se refieren a goles** marcados se puede considerar que **hay normalización**. En el **caso general habría que normalizar** las variables **si se quisiera comparar valores absolutos** entre distintas parejas.

## 7.-

```python
from sklearn.preprocessing import LabelEncoder
data=pd.read_excel('datos/all-euro-data-2020-2021.xlsx',sheet_name='SP1')
feats=['Div', 'HomeTeam' , 'AwayTeam']
encoder_liga=LabelEncoder()
encoder_teams=LabelEncoder()
data['Div']=encoder_liga.fit_transform(data['Div'].values)
teams=np.concatenate([data['HomeTeam'].values,data['AwayTeam'].values]) #Juntamos todos los equipos

encoder_teams.fit(teams)
data['HomeTeam']=encoder_teams.transform(data['HomeTeam'])
data['AwayTeam']=encoder_teams.transform(data['AwayTeam'])
print(data[feats].head())
```

## 8.-

```python
from sklearn.preprocessing import StandardScaler
feats=['HTHG','HTAG','FTHG','FTAG']
scaler=StandardScaler()
data[feats]=scaler.fit_transform(data[feats])
print(data[feats].head())
print()
data[feats].cov()
```

Esta es la matriz de **covarianza antes de normalizar**:

HTHG	HTAG	FTHG	FTAG
HTHG	0.607612	-0.010381	0.620069	-0.045675
HTAG	-0.010381	0.526488	-0.022074	0.544028
FTHG	0.620069	-0.022074	1.380217	-0.105548
FTAG	-0.045675	0.544028	-0.105548	1.094392

   - Los **signos no cambian**. Las **correlaciones, de forma cualitativa, no han cambiado.
    
   - **Cambian las magnitudes**, y **ahora sí son comparables** al ser variables normalizadas.
   
   - **La varianza es la misma para todas las variable**.

## 9.-

```python
def myfunction(goles_h,goles_a):
    if goles_h>goles_a:
        result='H'
    elif goles_h<goles_a:
        result='A'
    else:
        result='D'
    return result
```
