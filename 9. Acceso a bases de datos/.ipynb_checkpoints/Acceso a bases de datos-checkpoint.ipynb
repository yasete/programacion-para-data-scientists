# Acceso a bases de datos

# Librería pyodbc

La librería pyodbc **utiliza conexiones odbc para conectarse de forma estándar a DBMSs**. Pero primero **aclaremos los conceptos**.

# Definiciones:

## DBMS (Database Management System)

Es un **conjunto de programas que se encargan de manejar la creación y todos los accesos a las bases de datos**.

## API (Application Programming Interface)

Uno de los principales **propósitos** de una API consiste en **proporcionar un conjunto de funciones de uso general**, por ejemplo, para dibujar ventanas o iconos en la pantalla. De esta forma, **los programadores se benefician de las ventajas del API** haciendo uso de su funcionalidad, **evitándose el trabajo de programar todo desde el principio**.

## ODBC (Open Database Connectivity)

**ODBC es una especificación para una API de base de datos**. Esta **API** es **independiente de un DBMS o sistema operativo**. La API de ODBC es **independiente también del lenguaje**. 

**Las aplicaciones (en nuestro caso un script python) llaman a las funciones de estos controladores para tener acceso a los datos de una manera independiente del DBMS**. 

**Microsoft proporciona un administrador de controladores para los equipos que ejecutan Microsoft Windows**. Sin embargo hay controladores **ODBC desarrollados por otras empresas**. 

Además, los **controladores y las aplicaciones ODBC existen en los Mac y en** una variedad de **plataformas de UNIX**.

# 2. Instalación del controlador si no está en el administrador

En nuestro caso vamos a trabajar con el controlador de BigQuery, que es una base de datos de alto rendimiento desplegado en la nube de Google (Google Cloud Platform).

**Instalación del controlador de BigQuery en Windows**

https://cloud.google.com/bigquery/docs/reference/odbc-jdbc-drivers

# 3. Configurar el controlador en la pestaña DSN (Data Resource Name)

Se han de **configurar las características de la conexión** a la base de datos externa.

**Parámetros importantes para una conexión a BigQuery:**

   - Seleccionar **'Service Authentication' o 'User authentication'**
   
   - **'Key file path'**: Es el **path en nuestra máquina en el que guardamos el fichero .json** que hemos **descargado de GCP** para autenticación. Nuestras queries no pueden acceder sin esta autenticación.
   
   - **Project**:project que hemos creado en GCP.
   
   - **Dataset**: Aunque el dataset se puede especificar en el programa python, se puede configurar también aquí.

# 4. Escribir el programa que accede a la base de datos:

En este ejemplo vamos a acceder al **programa público de datasets de GCP** que provee acceso gratuito a multitud de datos sobre diferentes temas. En concreto vamos a acceder a un dataset donde se guardan datos de **evolución del Covid-19 por paises**.

```python
import pyodbc
import pandas as pd


#Ahora creamos el objeto conexión pasándole como parámetro el DSN, que es la configuración que ya hemos realizado antes para el controlador de BigQuery
cnxn = pyodbc.connect('DSN=Google BigQuery', autocommit=True) 
#cursor = cnxn.cursor()


#SQL query directamente desde pandas:
df=pd.read_sql_query('select * from bigquery-public-data.covid19_ecdc_eu.covid_19_geographic_distribution_worldwide',con=cnxn,parse_dates='date')

cnxn.close() #Cerramos la conexión
```

# 5. Realizar el análisis de los datos capturados

```python
print(df.head())
df.columns
```

```python
filtro=df[df['countries_and_territories']=='Spain']
filtro.index=filtro['date']
filtro=filtro.sort_index()
filtro
```
