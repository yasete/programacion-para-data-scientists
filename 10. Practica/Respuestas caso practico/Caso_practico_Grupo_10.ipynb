PREGUNTA #3

Creando funcion para hacer la descomposición en trend, season y residuo. Eliminando nulos dentro de la función
Ploteando trend, season y residuo dentro de la función.

```python
def decomp(TempSerie, frec=7):
  import matplotlib.pyplot as plt
  from statsmodels.tsa.seasonal import seasonal_decompose as decompose
  result=decompose(TempSerie, freq=7) 
  trend=result.trend
  season=result.seasonal
  residuo=result.resid
  trend=trend.dropna()
  season=season.dropna()
  residuo=residuo.dropna()
  componentes=[trend, season, residuo]
  plt.figure(figsize=(7,7))
  plt.subplot(311)
  plt.plot(trend)
  plt.subplot(312)
  plt.plot(season)
  plt.subplot(313)
  plt.plot(residuo)
  return trend, season, residuo
```

```python
from matplotlib import pyplot as plt
from statsmodels.tsa.stattools import adfuller as df
pvalues=[]#AÑADIDO X JUAN
result=df(residu,autolag='AIC')
print(result)
print()
print('Estadístico',result[0])
pvalues.append(result[1])
print('P-value',result[1])
print('Valores críticos',result[4])
print()
plt.plot(pvalues) 
plt.grid()
plt.show()
```
