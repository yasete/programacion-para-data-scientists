Importamos todos los módulos que necesitamos para realizar la práctica.

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.seasonal import seasonal_decompose as decompose
from statsmodels.tsa.stattools import adfuller as df
import matplotlib.pyplot as plt
%matplotlib inline
```

### 1. Importación de los datos

En primer lugar importamos los datos, que deben situarse en la misma ruta que este código. Como nos interesa únicamente la primera hoja, con el parámetro sheet_name indicamos el nombre de la misma, en este caso 'E0' que es la primera división inglesa. Otra forma sería con sheet_name=0 que coge la primera hoja del archivo excel sin importar su nombre.

```python
datos_original = pd.read_excel('all-euro-data-2020-2021.xlsx',sheet_name='E0')
datos_original.head() # Mostramos las 5 primeras filas para una inspeccion visual rapida
```

Seleccionamos las variables que nos interesan: Date, HTHG, HTAG, FTHG y FTAG.

```python
variables = ['Date', 'HTHG', 'HTAG', 'FTHG', 'FTAG']
datos = datos_original[variables]
datos.head()
```

En el enunciado nos piden transformar la variable Date al formato datetime, pero como se puede apreciar en la siguiente ejecución, al importar los datos la variable ya tiene dicho formato:

```python
datos.dtypes
```

### 2. Formamos una serie temporal en formato pd.Series

En primer lugar agrupamos por fecha, resumiendo el resto de variables con la media:

```python
datos_agrupados_fecha = datos.groupby(by='Date').mean()
datos_agrupados_fecha.head()
```

Creamos una Serie de Pandas con la variable de goles del equpo local al final del partido (FTHG)

```python
serie_fthg = pd.Series(datos_agrupados_fecha['FTHG'])
serie_fthg.head()
```

Comprobamos que realmente sea un objeto de tipo Series de Pandas:

```python
type(serie_fthg)
```

### 3. Creamos una función que haga la descomposición trend, seasonal y residuo

```python
def descomposicion(ts):
    
    result = decompose(ts,freq=7) 
    
    trend = result.trend
    season = result.seasonal
    resid = result.resid
    
    # Generamos los correspondientes gráficos:
    
    plt.figure(figsize=(10,10))
    
    plt.subplot(311)
    plt.plot(trend)
    plt.title('Tendencia')
    
    plt.subplot(312)
    plt.plot(season)
    plt.title('Estacionalidad')
    
    plt.subplot(313)
    plt.plot(resid)
    plt.title('Residuo')
    
    plt.tight_layout()
    plt.show()
    
    # Eliminamos los nulos
    
    trend.dropna(inplace = True)
    season.dropna(inplace = True)
    resid.dropna(inplace = True)
    
    return trend, season, resid
```

### 4. Aplicamos la función a la variable 'FTHG'

Gráfico de la serie antes de realizar ninguna descomposición:

```python
plt.figure(figsize=(10,3))
serie_fthg.plot()
plt.ylabel('FTHG')
plt.title('Serie original')
plt.show()
```

Aplicamos la función que hemos creado antes para realizar la descomposición y graficar la tendencia, la estacionalidad y el residuo

```python
tendencia, estacionalidad, residuo = descomposicion(serie_fthg)
```

### 5. Aplicamos el test Dickey-Fuller (df) al residuo para ver si es estacionario

```python
resultado_test = df(residuo, autolag='AIC')

print(resultado_test)
print()
print('Estadístico',resultado_test[0])
print('P-value',resultado_test[1])
print('Valores críticos',resultado_test[4])
print()
```

Observando los resultados podemos decir que el residuo de nuestra serie tras la descomposición es **ESTACIONARIO**. Se puede ver de 2 formas:

1) El estadístico es menor que el valor crítico del 1% (intervalo de confianza del 99%).

2) El p-valor es menor que 0.01.

Ambos resultados, nos indican que a un nivel de significación del 1% se rechaza la hipótesis nula, por lo que la serie es estacionaria (la hipótesis nula es la NO estacionariedad).

*Nota: Obviamente, al verificarse la estacionariedad en un intervalo de confianza del 99%, se verifica para los intervalos de 95% y 90%, ya que estos dos últimos son menos restrictivos.*

El ajuste de un modelo ARIMA sobre una serie estacionaria se usa normalmente para realizar predicciones a futuro de dicha serie, por lo que si la serie es de algún deporte, se puede plantar elaborar un método de apuestas en ese deporte a partir de la predicción resultante.
