```python
import pandas as pd
```

Pregunta 1

```python
df = pd.read_excel('all-euro-data-2020-2021.xlsx', sheet_name='E0')
df = df[['Date', 'HTHG', 'HTAG', 'FTHG', 'FTAG']]
pd.to_datetime(df['Date'])
print(df.dtypes)
```

Pregunta 2

```python
df_date = df.groupby(by='Date').mean()

series_fthg = pd.Series(df_date['FTHG'])
print(f"Series FTHG dtype: {type(series_fthg)}")
# print(f"Series: \n{series_fthg}")
```

Pregunta 3

```python
from statsmodels.tsa.seasonal import seasonal_decompose as season

# Metemos solo un parametro y el parámetro freq=7 lo metemos unicamente en la función season
def decompose(df_series):
  out = season(df_series, freq=7)

  trend = out.trend.dropna()
  seasonal = out.seasonal.dropna()
  resid = out.resid.dropna()

  trend.plot(legend=True, label="trend")
  seasonal.plot(legend=True, label="seasonal")
  resid.plot(legend=True, label="resid")

  return trend, seasonal, resid
```

Pregunta 4

```python
trend, seasonal, resid = decompose(series_fthg)
```

Pregunta 5

Vamos a realizar el test considerando que la hipótesis nula es no estacionaria. Para aceptar esta hipótesis el estadístico ha de ser mayor que los valores críticos. Además, también podemos aceptar esta hipótesis nula si el p-valor es mayor que 0.05 (95% de confianza). 

```python
from statsmodels.tsa.stattools import adfuller as df

result = df(resid)
print(f'Estadistico FTHG {result[0]}')
print(f'P-valor FTHG {result[1]}')
print(f'Valores criticos FTHG {result[4]}')
```

Como podemos ver, el p-valor es menor que 0.05, lo que nos hace ver que la serie es estacionaria, pues rechazamos nuestra hipotesis. Además, vemos también que el estadístico es menor que los valores críticos, por lo que por este estudio también rechazamos la hipótesis nula.

Pregunta 6

```python
from statsmodels.tsa.seasonal import seasonal_decompose as season
from statsmodels.tsa.stattools import adfuller as df

series_ftag = pd.Series(df_date['FTAG'])
print(f"Series dtype: {type(series_ftag)}")
# print(f"Series: \n{series_ftag}")

trend, seasonal, resid = decompose(series_ftag)

result = df(resid)
print()
print(f'Estadistico FTAG {result[0]}')
print(f'P-valor FTAG {result[1]}')
print(f'Valores criticos FTAG {result[4]}')
```

Como en el caso de FTHG, ahora también tenemos el p-valor menor que 5%, por lo que también rechazamos la hipótesis nula y determinamos que esta serie es estacionaria.

Pregunta 7

```python
from statsmodels.tsa.seasonal import seasonal_decompose as season
from statsmodels.tsa.stattools import adfuller as df

# HTHG
series_hthg = pd.Series(df_date['HTHG'])
print(f"Series dtype: {type(series_hthg)}")
# print(f"Series: \n{series_hthg}")

trend, seasonal, resid = decompose(series_hthg)

result = df(resid)
print()
print(f'Estadistico HTHG {result[0]}')
print(f'P-valor HTHG {result[1]}')
print(f'Valores criticos HTHG {result[4]}')
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose as season
from statsmodels.tsa.stattools import adfuller as df

# HTAG
series_htag = pd.Series(df_date['HTAG'])
print(f"Series dtype: {type(series_htag)}")
# print(f"Series: \n{series_htag}")

trend, seasonal, resid = decompose(series_htag)

result = df(resid)
print()
print(f'Estadistico HTAG {result[0]}')
print(f'P-valor HTAG {result[1]}')
print(f'Valores criticos HTAG {result[4]}')
```

Como podemos ovbservar, también en estos dos casos estudiados se observa que el p-valor es menor que el 5%, entonces vemos que ambas series también son estacionarias.

### **5. ¿Se te ocurre alguna forma de relacionar el concepto de estacionariedad en series temporales deportivas y la creación de un método de apuestas?**

Las series temporales (la estacionariedad) pueden ser utilizadas para la creación de un método de apuestas puesto que pueden ayudar a la hora de "predecir" posibles resultados al permitir un análisis del comportamiento más o menos cíclico de las variables. Es decir, a mayor estacionariedad de los datos, más confianza a la hora de apostar.
