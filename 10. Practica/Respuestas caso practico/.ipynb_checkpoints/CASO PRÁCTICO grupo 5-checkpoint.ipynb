```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose as season
from statsmodels.tsa.stattools import adfuller as fuller
```

## Importación de los datos

```python
##Cambiar el path local donde se encuentra descargado el documento
data=pd.read_excel ('all-euro-data-2020-2021.xlsx', sheet_name= 'E0')
```

**Escogemos los campos relevantes**

```python
variables=['Date','HTHG','HTAG', 'FTHG','FTAG']
df=data[variables]
df
```

**Convertimos 'Date' a formato 'Datetime'**

```python
## función astype
df.astype({'Date':'datetime64[ns]'})
```

```python
df.dtypes
```

**Agregamos la información por 'Date', promediando valores**

```python
df_mean=df.groupby('Date').mean()
```

```python
df_mean
```

**Formamos las series temporables de las variables**

```python
ts_fthg=pd.Series(df_mean['FTHG'])
ts_ftag=pd.Series(df_mean['FTAG'])
ts_hthg=pd.Series(df_mean['HTHG'])
ts_htag=pd.Series(df_mean['HTAG'])
```

# Funciones

Para la descomposición en [trend, seasonal, resid] creamos la función **decompose**, desde donde llamaremos a la función **viz**, que se encarga de graficar las señales.
 
Además, también hemos creado la función **resid_fuller** que aplica el **Test de Dickey-Fuller** a los residuos obtenidos en la función decompose y devuelve el estadístico, p-value y valores críticos.

El test de Dickey-Fuller considera como **hipótesis nula** que **la serie es no estacionaria**.

   - **Para aceptar la hipótesis nula** (serie no estacionaria) **el estadístico** (primer valor del resultado) **ha de ser mayor que los valores críticos**.

   - **Otro criterio** para acepatar la hipótesis nula es que el **p-value sea mayor que 0.05** (95% de confianza).

**JUAN:**
La función decompose tenía una cosa mal. season(ts,period=7), en realidad ese parámetro no existe, el parámetro al que os referis es 'freq'. Lo he modificado en la función para que el código no de error.
Equivocarse en el nombre del parámetro no es un problema en mi opinión, se parece en cuanto al signifcado al correcto, sin embargo sí considero importante que no hayáis comprobado que el código no da errores. Con la definición original la función da error y dicho error indica que el parámetro 'period' no existe.

```python
def viz (ts,trend,seasonal,resid):
    plt.figure(figsize=(10,15))
    plt.subplot(411)
    plt.title('Serie original')
    plt.plot(ts)
    plt.subplot(412)
    plt.title('Tendencia')
    plt.plot(trend, color='g')
    plt.subplot(413)
    plt.title('Estacional')
    plt.plot(seasonal, color='y')
    plt.subplot(414)
    plt.title('Residuo')
    plt.plot(resid, color='m')
    
def decompose(ts):
    aux=season(ts,freq=7)
    trend=aux.trend.dropna()
    seasonal=aux.seasonal.dropna()
    resid=aux.resid.dropna()
    viz(ts,trend,seasonal,resid)
    return [trend,seasonal,resid]

def resid_fuller(ts):
    decomposing=decompose(ts)
    resid=decomposing[2]
    result=fuller(resid,autolag='AIC')
    print(result)
    print()
    print('Estadístico',result[0])
    print('P-value',result[1])
    print('Valores críticos',result[4])
    print()
```

**Ejecutamos las funciones que hemos creado arriba mediante un bucle for**

```python
series=[ts_fthg,ts_ftag,ts_hthg,ts_htag]
for i in series:
        print ('                                               ',i.name)
        print('**Decomposing**')
        print(decompose(i))
        print ('          ')
        print ('**Test Dickey-Fuller//Resid**')
        resid_fuller(i)
```

## Interpretación de los resultados

En los cuatro casos, el estadístico es menor que el valor crítico del 1% (99%) de confianza, así como el p-value, que se sitúa por debajo de 0.05 en todos los sucesos estudiados. De esta forma, podemos rechazar la hipótesis nula y considerar los residuos de las cuatro series como **estacionarios** (media y varianza constante).

*2. ¿Encuentras alguna justificación para las diferencias encontradas?. ¿Hay algún aspecto del juego que justifique aparentemente dichas diferencias?*

Las circunstacias al final de un partido pueden variar mucho de las que se dan en el descanso. En primer lugar, los jugadores están más cansados, algo que también marca el ritmo del partido en cuanto a rapidez en el juego y número de jugadas, lo que hace que aumente la tendencia y la estacionalidad.

Otro aspecto podría ser que el resultado ya esté determinado a mitad de tiempo. En este sentido, tener un marcador definido ya en la segunda parte del partido, puede desencadenar en jugadas sin mucha variabilidad, es decir, la media y la desviación estándar se mantendrían constantes en el tiempo. 

Por último, la calidad de los jugadores también puede influir. Generalmente, en la segunda parte del partido o últimos minutos, algunos titulares suelen ser sustituidos, bien por descanso, o bien porque no está haciendo el partido que se espera.

**JUAN: La clave en realidad es el tiempo transcurrido. En un proceso en el tiempo hay una serie de factores potenciales que se manifestarán, en termino medio, a lo largo de dicho proceso. Cuanto más tiempo le demos al proceso más oportunidades le damos a los factores potenciales de manifestarse. Cuando el tiempo es muy corto hay mucha más variabilidad y se producen los llamados rallies (en bolsa ocurre por ejemplo) que son señales ruidosas que nada tienen que ver con la dinámica real. Sin embargo cuando al proceso se le deja más tiempo tiende poco a poco a los valores correctos que rigen su dinámica.**

Un ejemplo en el que podéis verlo es si simuláis una serie con numpy que sea por ejemplo una distribución normal con media mu y varianza var. Si simuláis muy pocos valores y calculáis el valor medio y su varianza, seguramente esos valores calculados estarán muy lejos de la muy y el var que habéis puesto en la simulación. Sin embargo si seguís simulando más y más valores veréis como poco a poco la simulación tiende a los valores que habéis programado en la función de numpy.

En resumen, a periodos más cortos se produce más ruido o variabilidad, lo que hace que la serie no manifieste su dinámica real.
