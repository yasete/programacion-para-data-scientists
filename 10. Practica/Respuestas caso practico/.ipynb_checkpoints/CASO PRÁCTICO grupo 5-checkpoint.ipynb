# Funciones

Para la descomposici√≥n en [trend, seasonal, resid] creamos la funci√≥n **decompose**, desde donde llamaremos a la funci√≥n **viz**, que se encarga de graficar las se√±ales.
 
Adem√°s, tambi√©n hemos creado la funci√≥n **resid_fuller** que aplica el **Test de Dickey-Fuller** a los residuos obtenidos en la funci√≥n decompose y devuelve el estad√≠stico, p-value y valores cr√≠ticos.

El test de Dickey-Fuller considera como **hip√≥tesis nula** que **la serie es no estacionaria**.

   - **Para aceptar la hip√≥tesis nula** (serie no estacionaria) **el estad√≠stico** (primer valor del resultado) **ha de ser mayor que los valores cr√≠ticos**.

   - **Otro criterio** para acepatar la hip√≥tesis nula es que el **p-value sea mayor que 0.05** (95% de confianza).

**JUAN:**
La funci√≥n decompose ten√≠a una cosa mal. season(ts,period=7), en realidad ese par√°metro no existe, el par√°metro al que os referis es 'freq'. Lo he modificado en la funci√≥n para que el c√≥digo no de error.
Equivocarse en el nombre del par√°metro no es un problema en mi opini√≥n, se parece en cuanto al signifcado al correcto, sin embargo s√≠ considero importante que no hay√°is comprobado que el c√≥digo no da errores. Con la definici√≥n original la funci√≥n da error y dicho error indica que el par√°metro 'period' no existe.

```python
def viz (ts,trend,seasonal,resid):
    plt.figure(figsize=(10,15))
    plt.subplot(411)
    plt.title('Serie original')
    plt.plot(ts)
    plt.subplot(412)
    plt.title('Tendencia')
    plt.plot(trend, color='g')
    plt.subplot(413)
    plt.title('Estacional')
    plt.plot(seasonal, color='y')
    plt.subplot(414)
    plt.title('Residuo')
    plt.plot(resid, color='m')
    
def decompose(ts):
    aux=season(ts,freq=7)
    trend=aux.trend.dropna()
    seasonal=aux.seasonal.dropna()
    resid=aux.resid.dropna()
    viz(ts,trend,seasonal,resid)
    return [trend,seasonal,resid]

def resid_fuller(ts):
    decomposing=decompose(ts)
    resid=decomposing[2]
    result=fuller(resid,autolag='AIC')
    print(result)
    print()
    print('Estad√≠stico',result[0])
    print('P-value',result[1])
    print('Valores cr√≠ticos',result[4])
    print()
```

**Ejecutamos las funciones que hemos creado arriba mediante un bucle for**

```python
series=[ts_fthg,ts_ftag,ts_hthg,ts_htag]
for i in series:
        print ('                                               ',i.name)
        print('**Decomposing**')
        print(decompose(i))
        print ('          ')
        print ('**Test Dickey-Fuller//Resid**')
        resid_fuller(i)
```

## Interpretaci√≥n de los resultados

En los cuatro casos, el estad√≠stico es menor que el valor cr√≠tico del 1% (99%) de confianza, as√≠ como el p-value, que se sit√∫a por debajo de 0.05 en todos los sucesos estudiados. De esta forma, podemos rechazar la hip√≥tesis nula y considerar los residuos de las cuatro series como **estacionarios** (media y varianza constante).
