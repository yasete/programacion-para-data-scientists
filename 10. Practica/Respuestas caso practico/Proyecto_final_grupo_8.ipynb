```python
import pandas as pd
import numpy as np
from datetime import date, datetime
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
data = pd.read_excel(io='all-euro-data-2020-2021.xlsx', sheet_name= 'SP1')
data.head()
```

```python
# Convertimos la variable Date a formato datetime
data['Date'] = pd.to_datetime(data['Date'])
```

EJERCICIO 2

```python
# Se agrega la información en un nuevo dataframe pandas usando groupby, y agregando por variable 'Date', 
#promediando los valores del resto de variables.

df_grouped= data.groupby('Date').mean()
```

```python
# Se elige la variable ' FTHG' y se forma una serie temporal

fthg_series = pd.Series(df_grouped['FTHG'])
```

```python
# Se comprueba que el resultado es pd.Series.
type(fthg_series)
```

```python
fthg_series 
```

Test Dickey Fuller

```python
window=7
```

```python
from statsmodels.tsa.stattools import adfuller as df

n_tramos=int(round(len(fthg_series)/window,0)) #Número de tramos

pvalues=[]

for cont in range(0,n_tramos): 
    print('Tramo',cont)
    ts_temp=fthg_series[cont*window:(cont+1)*window]
    result=df(ts_temp,autolag='AIC')
    print(result)
    print()
    print('Estadístico',result[0])
    pvalues.append(result[1])
    print('P-value',result[1])
    print('Valores críticos',result[4])
    print()
plt.plot(pvalues)
plt.grid()
plt.show()
    
```

**JUAN: Habéis cometido un error grave al llamar a vuestra función de la misma forma que el alias que le dais a  seasonal_decompose un poco más abajo (from statsmodels.tsa.seasonal import seasonal_decompose as decompose). Lo que ocurre es que al llamar más abajo seasonal_decompose como decompose estáis machando vuestra función, y cuando uséis ese alias estáis usando seasonal_decompose en vez de vuestra función (que por otro lado en este ejemplo hacen lo mismo). Como norma general hay que tener cuidado y no dar los mismos nombres a funciones y alias.**

```python
# Creamos una función que haga la descomposición trend, seasonal y residuo.
# Dicha función ha de hacer la descomposición en tres señales (trend, seasonal y residuo) de una serie temporal.
# Dentro de la función eliminad los nulos generados en la descomposición antes de devolver las señales a la salida.

def decompose(ts, freq=window):

    from statsmodels.tsa.seasonal import seasonal_decompose as decompose


    
    result=decompose(fthg_series, freq= window) 
    trend=result.trend
    season=result.seasonal
    residuo=result.resid
    
    trend= trend.dropna()
    season= season.dropna()
    residuo= residuo.dropna()
    
    plt.figure(figsize=(20,20))
    plt.subplot(311)
    plt.plot(trend)
    plt.subplot(312)
    plt.plot(season)
    plt.subplot(313)
    plt.plot(residuo)
          
    return trend, season, residuo

decompose(fthg_series)
```
