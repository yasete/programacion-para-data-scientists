## 1.- Importa el fichero de datos covid_19_complete.xls

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
covid=pd.read_csv("datos/covid_19_complete.csv")
```

## 2.- Realiza la limpieza necesaria de nulos, imputando la media de su columna.

```python
print(covid.info())
print(covid.describe())
```

```python
covid.isnull()
```

```python
covid=covid.fillna(covid.mean())
covid.isnull()
```

## 3.- Descubre por qué algunos nulos no desaparecen.

Los nulos que no desaparecen es porque pertenecen a una variable no numérica que no admite cálculo de la media en la columna.

Procedemos a sustituir esos nulos en la variable Province/state por el valor 'Null-land'

## 4.- Trata dichos nulos de la manera que consideres necesaria pero hazlos desaparecer.

```python
covid=covid.fillna('Null-land')
covid.isnull()
```

```python
covid.head(10)
```

## 5.- Dibuja un mapa de calor de la matriz de correlación del dataframe.

```python
#PARA HACER LA MATRIZ DE CORRELACIÓN NECESITAMOS AISLAR LAS VARIABLES NUMÉRICAS
feats=covid.columns
feats_num=[]
for feat in feats:
    if (covid[feat].dtype=='float64') or (covid[feat].dtype=='int64'):
        feats_num.append(feat)
covid_num=covid[feats_num]
covid_num.head()
```

```python
#HEATMAP
sns.heatmap(covid_num.corr(),cmap='coolwarm')
```

## 6.- Haz un mapa de clustering sobre dicho mapa de calor.

```python
#CLUSTERING
sns.clustermap(covid_num.corr(),cmap='coolwarm')
```

Claramente hay una correlación entre la variable 'casos confirmados' y la variable 'deaths'.Es obvio que el crecimiento de la primera está asociado con un crecimiento de la segunda.

Por otro lado no se observa correlación entre casos confirmados y muertes con latitud/longitud.

## 7.- Realiza un gráfico de regresión entre las dos variables con mayor correlación.

```python
sns.lmplot(x='Confirmed',y='Deaths',data=covid_num)
```

Claramente aunque hay correlación un modelo lineal con solo una variable no representa bien muertes como función de casos

## 8.- Dibuja una distribución de probabilidad de la variable 'Death' por pais

```python
#AGRUPAMOS PRIMERO POR PAIS
covid_pais=covid.groupby(by='Country/Region').sum()
covid_pais.head()
```

```python
#DISTRICUCIÓN DE PROBABILIDAD
sns.set_style('dark')
sns.kdeplot(covid_pais['Deaths'],color='blue')
```

```python
covid_filter=covid_pais[covid_pais.Deaths>40000]
covid_filter.head(14)
```

## 9.- Realiza dos gráficos de barras dentro del mismo área, los dos correspondiendo a los 5 paises con mayor y menor número de casos respectivamente.

```python
paises_mas_casos=covid_pais.sort_values(by='Confirmed',ascending=False) #sort en orden descendente
paises_mas_casos=paises_mas_casos.iloc[0:5] #Seleccionamos los cinco primeros

#Ahora hacemos lo mismo para los de menos casos
paises_menos_casos=covid_pais.sort_values(by='Confirmed',ascending=True) #sort en orden ascendente
paises_menos_casos=paises_menos_casos.iloc[0:5] #Seleccionamos los cinco primeros
```

```python
#Dibujamos los gráficos de barras.

plt.subplot(1,2,1)
plt.ylabel('Casos')
plt.title('Paises con más casos')
plt.xticks(rotation='vertical')
plt.bar(paises_mas_casos.index,paises_mas_casos.Confirmed)
plt.subplot(1,2,2)
plt.ylabel('Casos')
plt.title('Paises con menos casos')
plt.xticks(rotation='vertical')
plt.bar(paises_menos_casos.index,paises_menos_casos.Confirmed)
plt.show()
```
