We will catch the error using the **try: ... except** statements.
Try-except: you  specify a block of code for the computer to try and execute, if it encounters an error, instead of failing to run the whole syntax, it goes to a blocl of code you've provided in case of an error. This is called Error Catching


```python
# without error catching
my_text = "Text"
print(int(my_text))
```

```python
# with error catching
my_text = "not a number"
try:
    a = int(my_text) # in this example we will never go here
    print("The value of the text is", a)
except:
    a = 0 # instead we will go here
    print("The value cannot be converted to integer")
```

```python
my_text = "100"
try:
    a = int(my_text) # in this example we will never go here
    print("The value of the text is", a)
except:
    a = 0 # instead we will go here
    print("The value cannot be converted to integer")
```

Lets implement the **try-except** when we read teh data. We will replace all strings that cannot be converted to integers with **None** which is a special value meaning a missing value.

```python
rows = []
f = open("MovieData.csv")
header = f.readline()
print(header)
for i in f: # this will start from the second row
    listn = i.split('\t') 
    try:
        listn[3] = int(listn[3])
    except:
        listn[3] = None
    try:
        listn[4] = int(listn[4])
    except:
        listn[4] = None
    try:
        listn[5] = int(listn[5])
    except:
        listn[5] = None

    rows.append(listn)
```

```python
rows[2] # check one of the troublesome rows
```

That was a lot of typing though. Lets wrap the **try-except** into a function

```python
def convert_into_int(x):
    try:
        return int(x)
    except:
        return None
```

```python
rows = []
f = open("MovieData.csv")
header = f.readline()
for i in f: # this will start from the second row
    listn = i.split('\t') 
    listn[3] = convert_into_int(listn[3])
    listn[4] = convert_into_int(listn[4])
    listn[5] = convert_into_int(listn[5])
    rows.append(listn)
rows[1000]
```

Again we have 3 lines that look the same. Lets replace them with a loop

```python
rows = []
f = open("MovieData.csv")
header = f.readline()
for i in f: # this will start from the second row
    listn = i.split('\t') 
    for r in range(3,6):
        listn[r] = convert_into_int(listn[r])
    rows.append(listn)
rows[1000]
```

Write a conversion function

```python
def make_date(date_str):
    ''' turn DD/MM/YY object into a datetime object'''
    m, d, y = date_str.split('/')
    m = m.strip('"')
    m = int(m)
    d = int(d)
    y = int(y)
    if y > 13:
        y += 1900
    else: 
        y += 2000
    return dt.datetime(y, m, d)
```

Now lets integrate the function into the date reading

```python
rows = []
f = open("MovieData.csv")
header = f.readline()
for i in f: # this will start from the second row
    listn = i.split('\t') 
    listn[0] = make_date(listn[0]) # convert the date into string into a datetime object
    for r in range(3,6):
        listn[r] = convert_into_int(listn[r])
    rows.append(listn)
print(rows[1000])
```

# List comprehension

We want to do analysis column-wise. To do this we need to create new lists of only the rows we want - think of it as selecting columns

```python
all_dates = []
for d in rows:
    all_dates.append(d[0])
print(len(all_dates)) # number of records
print(all_dates[0]) # first record
```

There is a specific idiom for creating lists from other lists, called **list comprehension**. It puts the for loop from above in square brakets

```python
all_dates = [d[0] for d in rows]
print(len(all_dates)) # number of records
print(all_dates[0]) # first record
```

Now that we have a list of dates we can use the build in functions

```python
print(min(all_dates)) # earliest date
print(max(all_dates)) # latest date
```

Lets analyse the budgets

```python
budgets = [b[3] for b in rows]
print(len(budgets))
print(max(budgets))
```

```python
def find_mean(num_list):
    '''
    Find the average of num_list
    '''
    total = 0.0
    for x in num_list:
        total += x
    return total / len(num_list)
find_mean(budgets)
```
